```python
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 读取两个CSV文件
df_a = pd.read_csv(r'D:\测试病例(100例)\labels.csv', dtype={'slide_id': str})  # 包含 slide_id, label（真实标签）
df_b = pd.read_csv(r'D:\测试病例(100例)\测试\测试_result\cancer\merged_voting_result.csv', dtype={'slide_id': str})

prefix = '/NAS3/lbliao/Data/MXB/classification/100例测试/feat_1_224/pt_files/h-optimus-1/'
suffix = '.pt'

# 清洗slide_id：先去掉前缀，再去掉后缀
df_a['slide_id'] = df_a['slide_id'].str.replace(prefix, '', regex=False)  # 去掉前缀（非正则模式，避免特殊字符干扰）
df_a['slide_id'] = df_a['slide_id'].str.replace(suffix, '', regex=False)  # 去掉后缀
df_b = df_b.drop('label', axis=1, errors='ignore')
df_a['slide_id'] = df_a['slide_id'].str.replace(r'-', '', regex=False).str.replace(r'.', '', regex=False)
df_b['slide_id'] = df_b['slide_id'].str.replace(r'-', '', regex=False).str.replace(r'.', '', regex=False)
df_a['slide_id'] = df_a['slide_id'].str.lower()
df_b['slide_id'] = df_b['slide_id'].str.lower()
# 按slide_id合并两个DataFrame，只保留两边都存在的ID
merged = pd.merge(df_a, df_b, on='slide_id', how='inner')

# 提取真实标签和预测标签
y_true = merged['label']
y_pred = merged['prediction']

# 计算评估指标
print("准确率（Accuracy）：", accuracy_score(y_true, y_pred))
print("\n混淆矩阵：")
print(confusion_matrix(y_true, y_pred))
print("\n分类报告（包含精确率、召回率、F1值等）：")
print(classification_report(y_true, y_pred))
```
